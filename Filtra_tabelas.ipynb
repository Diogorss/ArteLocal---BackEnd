BUSCA EM PLANILHAS NORMAIS


```python

import pandas as pd
from google.colab import files
import io
import os


# --- ÁREA DE CONFIGURAÇÃO PRINCIPAL ---
# Preencha todas as informações aqui antes de executar

# 1. ESPECIFIQUE O NOME DO SEU ARQUIVO DE ENTRADA:
NOME_DO_ARQUIVO_DE_ENTRADA = 'Levantamento de Trocadores de Calor pelo SDCD.xlsx'

# 2. ESPECIFIQUE A PLANILHA (ABA) QUE VOCÊ QUER USAR:
# Use o nome exato (ex: 'Visão Geral') ou o número da posição (0 para a primeira).
NOME_DA_PLANILHA_OU_INDICE = "Cronograma"

# 3. CONFIGURE OS FILTROS:
# FILTRO 1 (INCLUIR):
coluna_filtro_1 = 'Motivo para não implementar'
valor_filtro_1 = 'Falta Instrumentação'

# FILTRO 2 (INCLUIR):
coluna_filtro_2 = 'Motivo para não implementar'
valor_filtro_2 = 'Falta Instrumentação e  TAG Avariada'

# FILTRO 3 (EXCLUIR):
coluna_filtro_3 = 'Comentário'
valor_a_excluir_3 = 'Unidade hibernada'

# 4. CONFIGURE AS COLUNAS DE RETORNO:
colunas_de_retorno = ['Tag', 'Comentário']

# 5. NOME DO ARQUIVO DE SAÍDA:
nome_arquivo_saida = 'relatorio_filtrado.xlsx'

# --- FIM DA ÁREA DE CONFIGURAÇÃO ---


if not os.path.exists(NOME_DO_ARQUIVO_DE_ENTRADA):
    print(f"Arquivo '{NOME_DO_ARQUIVO_DE_ENTRADA}' não encontrado.")
    print("Por favor, faça o upload do arquivo agora.")
    uploaded = files.upload()

    if NOME_DO_ARQUIVO_DE_ENTRADA not in uploaded:
        print(f"\nERRO: Você fez o upload de um arquivo, mas o nome não corresponde a '{NOME_DO_ARQUIVO_DE_ENTRADA}'.")
        df = None
    else:
        print(f"\nArquivo '{NOME_DO_ARQUIVO_DE_ENTRADA}' carregado com sucesso!")
        df = pd.read_excel(io.BytesIO(uploaded[NOME_DO_ARQUIVO_DE_ENTRADA]), sheet_name=NOME_DA_PLANILHA_OU_INDICE)
else:
    print(f"Arquivo '{NOME_DO_ARQUIVO_DE_ENTRADA}' já encontrado. Carregando da planilha '{NOME_DA_PLANILHA_OU_INDICE}'...")
    df = pd.read_excel(NOME_DO_ARQUIVO_DE_ENTRADA, sheet_name=NOME_DA_PLANILHA_OU_INDICE)

if df is not None:
    print("\n--- Amostra dos seus dados (5 primeiras linhas) ---")
    print(df.head())
    print("\n----------------------------------------------------")

    print(f"\nFiltros aplicados:")
    print(f"1. Na coluna '{coluna_filtro_1}' com o valor '{valor_filtro_1}'")
    print(f"2. Na coluna '{coluna_filtro_2}' com o valor '{valor_filtro_2}'")
    print(f"3. Na coluna '{coluna_filtro_3}' EXCLUINDO o valor '{valor_a_excluir_3}'")
    print(f"Colunas que serão retornadas: {colunas_de_retorno}")
    print("----------------------------------------------------\n")

    try:

        df_filtrado = df[
    ( (df[coluna_filtro_1] == valor_filtro_1) | (df[coluna_filtro_2] == valor_filtro_2) ) &
    (df[coluna_filtro_3].astype(str).str.strip().str.lower() != valor_a_excluir_3.strip().lower())
]

        df_resultado = df_filtrado[colunas_de_retorno]

        if not df_resultado.empty:
            df_resultado.insert(loc=1, column='', value='')

        print("--- RESULTADO FILTRADO (será exportado para o novo arquivo) ---")
        if df_resultado.empty:
            print("Nenhum dado encontrado com os filtros especificados. Nenhum arquivo será gerado.")
        else:
            print(df_resultado)
            df_resultado.to_excel(nome_arquivo_saida, index=False, engine='openpyxl')
            print(f"\n--- EXPORTAÇÃO COMPLETA ---")
            print(f"O arquivo '{nome_arquivo_saida}' está pronto! Iniciando o download...")
            files.download(nome_arquivo_saida)

    except KeyError as e:
        print(f"\nERRO: Uma das colunas que você especificou não foi encontrada na planilha '{NOME_DA_PLANILHA_OU_INDICE}': {e}")
    except Exception as e:
        print(f"Ocorreu um erro inesperado: {e}")
```

MAPEAMENTOS


```python
##CÓDIGO PARA MAPEAMENTOS

import pandas as pd
from google.colab import files
import io
import os

# --- ÁREA DE CONFIGURAÇÃO PRINCIPAL ---

# 1. ESPECIFIQUE O NOME DO SEU ARQUIVO DE ENTRADA:
NOME_DO_ARQUIVO_DE_ENTRADA = 'LD Geral REFMAT  09092022.xlsx'

# 2. ESPECIFIQUE A PLANILHA (ABA) QUE VOCÊ QUER USAR:
NOME_DA_PLANILHA_OU_INDICE = 0

# 3. CONFIGURE OS FILTROS FIXOS E COLUNAS:
# O Filtro 1 (TAGs) será pedido ao usuário dentro do loop.

# FILTRO 2 (FIXO):
coluna_filtro_2 = 'NUMERO DOCUMENTO'
valor_filtro_2_fixo = 'FD' # Este valor será usado em todas as buscas

# COLUNA DO FILTRO 1 (que será variável):
coluna_filtro_1_variavel = 'TAGs'

# 4. CONFIGURE AS COLUNAS DE RETORNO:
colunas_de_retorno = ['NUMERO DOCUMENTO', 'TAGs', 'TITULO']

# 5. NOME DO ARQUIVO DE SAÍDA:
nome_arquivo_saida = 'mapeamento_filtrado.xlsx'

# --- FIM DA ÁREA DE CONFIGURAÇÃO ---


if not os.path.exists(NOME_DO_ARQUIVO_DE_ENTRADA):
    print(f"Arquivo '{NOME_DO_ARQUIVO_DE_ENTRADA}' não encontrado.")
    print("Por favor, faça o upload do arquivo agora.")
    uploaded = files.upload()
    if NOME_DO_ARQUIVO_DE_ENTRADA not in uploaded:
        print(f"\nERRO: O nome do arquivo enviado não corresponde a '{NOME_DO_ARQUIVO_DE_ENTRADA}'.")
        df = None
    else:
        print(f"\nArquivo '{NOME_DO_ARQUIVO_DE_ENTRADA}' carregado com sucesso!")
        df = pd.read_excel(io.BytesIO(uploaded[NOME_DO_ARQUIVO_DE_ENTRADA]), sheet_name=NOME_DA_PLANILHA_OU_INDICE)
else:
    print(f"Arquivo '{NOME_DO_ARQUIVO_DE_ENTRADA}' já encontrado. Carregando...")
    df = pd.read_excel(NOME_DO_ARQUIVO_DE_ENTRADA, sheet_name=NOME_DA_PLANILHA_OU_INDICE)

if df is not None:
    resultados_coletados = []

    print("\n" + "="*50)
    print("INÍCIO DA BUSCA INTERATIVA")
    print(f"Filtro fixo aplicado: Coluna '{coluna_filtro_2}' deve conter '{valor_filtro_2_fixo}'")
    print("="*50 + "\n")

    while True:
        valor_filtro_1_digitado = input(f"Digite o valor para a coluna '{coluna_filtro_1_variavel}' (ou 'sair' para finalizar): ")

        if valor_filtro_1_digitado.lower().strip() == 'sair':
            print("\nFinalizando a busca...")
            break

        try:

            df_filtrado_loop = df[
                (df[coluna_filtro_1_variavel].astype(str).str.contains(valor_filtro_1_digitado, case=False, na=False)) &
                (df[coluna_filtro_2].astype(str).str.contains(valor_filtro_2_fixo, case=False, na=False))
            ]

            if not df_filtrado_loop.empty:
                print(f"--- {len(df_filtrado_loop)} resultado(s) encontrado(s) para '{valor_filtro_1_digitado}'. Adicionando à lista final.")
                resultados_coletados.append(df_filtrado_loop)
            else:
                print(f"--- Nenhum resultado encontrado para '{valor_filtro_1_digitado}'.")

        except KeyError as e:
            print(f"\nERRO: Uma das colunas não foi encontrada: {e}")
            break
        except Exception as e:
            print(f"Ocorreu um erro inesperado: {e}")

        print("-" * 50)

    if not resultados_coletados:
        print("\nNenhum dado foi coletado. Nenhum arquivo será gerado.")
    else:

        df_final_combinado = pd.concat(resultados_coletados, ignore_index=True)

        df_final_combinado.drop_duplicates(inplace=True)

        df_resultado_final = df_final_combinado[colunas_de_retorno]

        if not df_resultado_final.empty:
            df_resultado_final.insert(loc=1, column='', value='')

        print("\n\n" + "="*50)
        print("RESULTADO FINAL COMPILADO")
        print("="*50)
        print(df_resultado_final)


        df_resultado_final.to_excel(nome_arquivo_saida, index=False, engine='openpyxl')
        print(f"\n--- EXPORTAÇÃO COMPLETA ---")
        print(f"O arquivo '{nome_arquivo_saida}' está pronto! Iniciando o download...")
        files.download(nome_arquivo_saida)
```
